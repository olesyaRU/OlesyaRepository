```python
import pandas as pd
from random import uniform, choice,seed
seed(42)
data1 = pd.DataFrame(data ={'a':[round(uniform(10, 20),2) for i in range(10)],
                            'b':[round(uniform(1, 2),2) for i in range(10)],
                            'c':[round(uniform(0, 3),2) for i in range(10)],
                            'sex':[choice(['f','m']) for i in range(10)]
                            })
data1
```

```python
seed(73)
data2 = pd.DataFrame(data ={'A':[round(uniform(10, 20),2) for i in range(10)],
                            'B':[round(uniform(1, 2),2) for i in range(10)],
                            'D':[round(uniform(4, 8),2) for i in range(10)],
                            'Sex':[choice(['F','M']) for i in range(10)]
                            })
data2
```


У нас есть данные. По заданию, вы должны определить, "какие колонки в этих двух датасетах соответствуют одним и тем же параметрам".

В колонках 'a' и 'A', 'b' и 'B', 'sex' и 'Sex' лежат одинаковые по смыслу значения, но с разными названиями столбцов.

Показатель 'с' есть только в data1, показатель 'D' есть только в data2.

Мы хотим соединить два датасета, чтобы увеличить выборку. Нужно отказаться от параметров 'с' и 'D'. Мы уменьшаем количество признаков в пользу увеличения количества объектов в выборке. Как правило, больший размер выборки - лучше, большего признакового пространства.

Увеличение размера выборки за счет объединения датасетов может помочь в обучении более точных моделей, особенно если исходный размер выборки небольшой. При этом отказ от части признаков может быть оправданным, если эти признаки либо не несут существенной информации для решаемой задачи, либо их наличие может внести шум или излишнюю сложность в модель.

В данной задаче у нас нет иного выхода, кроме как отказаться от части признаков, так как мы не можем "восстановить" значения для недостающих признаков.



```python
# Подготовка к объединению
data2['Sex'] = data2['Sex'].apply(lambda x: 'f' if x=='F' else 'm')
data2.rename(columns={'A':'a','B':'b', 'Sex':'sex'}, inplace=True)
data2
```

```python
data = pd.concat([data1,data2], ignore_index=True) # Соединяем датасеты
data
```

```python
data.drop(columns=['c', 'D'], inplace=True) # Удаляем признаки
data
```
